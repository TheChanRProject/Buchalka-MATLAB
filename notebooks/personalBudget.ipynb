## Personal Budget Per Day Calculator 

```matlab
% How much income do I have per week? 

income = 200 

% monthly subscriptions 

github = 7 
dropbox = 10
spotify = 10 
tinder = 15 
freelancer = 11 

% meal per day 

lunch = 10 
dinner = 10 

% costs per week 

weeklyCost = ((2*(2*lunch)) + (2*lunch) + (4*dinner) + (3*dinner))  

% costs per month 

monthlyCost = 4*weeklyCost + github + dropbox + spotify + tinder + freelancer 

% Leftover Spending Per Week 

weeklySpending = income - weeklyCost 

% Leftover Spending Per Month 

monthlySpending = (4*income) - monthlyCost 

disp(['I can spend ' num2str(monthlySpending / 30) ' per day ' 'if I spend twice on weekends than weekdays']) 



```
